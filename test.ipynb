```python
import os.path
from bs4 import BeautifulSoup
import requests
import csv
```

```python
def get_url(n):
    base_url = 'https://www.aldiwan.net'
    poem = f'poem{n}.html'
    return base_url + '/' + poem


def scrap_page(scrap_url):
    r = requests.get(scrap_url)
    if 'عذراً، الصفحة المطلوبة غير موجودة' in r.text:
        return None, None
    soup = BeautifulSoup(r.text, 'html.parser')
    
    poem_content = soup.find('div', {'id': 'poem_content'})
    
    text_content = poem_content.findAll('h3')
    
    text_list = [i.text.strip() for i in text_content if i.text]
    text = '\n'.join(text_list)
    
    tags_content = poem_content.find_next_siblings()[1].text.strip().split('\n')
    
    tags_list = [t for t in tags_content if t]
    tags = '-'.join(tags_list)
    
    return text, tags


def write_data(n, text, tags):
    filename = f'data/poems.csv'
    dirname = os.path.dirname(filename)
    if not os.path.isdir(dirname):
        os.makedirs(dirname)
    with open(filename, 'a', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow((n, text, tags))
```

```python
for n in range(22, 120_000):
    print(f'scraping {n}')
    text, tags = scrap_page(get_url(n))
    if text and tags:
        write_data(n, text, tags)
```
